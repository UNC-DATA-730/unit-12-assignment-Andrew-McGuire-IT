**Create and visualize your model below...**

````r
# Load the necessary libraries
library(ISLR2)
library(tidymodels)

# Define the recipe for data preprocessing
rec <- recipe(mpg ~ displacement + cylinders, data = Auto) %>%
  step_poly(displacement, degree = 2) %>%
  step_mutate(cyl_cat = factor(cylinders)) %>%
  step_dummy(cyl_cat) %>%
  step_rm(cylinders)

# Define the model specification (linear model)
model_spec <- linear_reg() %>%
  set_engine("lm")

# Create the workflow and fit the model
mod_fit <- workflow() %>%
  add_recipe(rec) %>%
  add_model(model_spec) %>%
  fit(data = Auto)

# Visualize the model
predictions <- predict(mod_fit, Auto) %>%
  bind_cols(Auto)

ggplot(predictions, aes(x = displacement, y = mpg)) +
  geom_point(size = 1, alpha = 0.35) +
  geom_line(aes(y = .pred, color = factor(cylinders)), linewidth = 1) +
  labs(title = "MPG as a Function of Displacement and Cylinders",
       x = "Displacement",
       y = "Miles per Gallon (MPG)",
       color = "Cylinders") +
  theme_minimal()
````
